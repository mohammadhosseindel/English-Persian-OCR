```python
import pytesseract
from pytesseract import Output
from PIL import Image
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# You Should Install Pytesseract SoftWare With Links In README.md File
# After Installing Pytesseract , Paste Pytesseract Installation Directory
pytesseract.pytesseract.tesseract_cmd = "Installation Directory Address/tesseract.exe"
```

```python
# Paste Image Address You Want To Extract Text
image = cv.imread("Image Address",0)
image_rgb = cv.imread("Image Address",0)
```

```python
# If Your Image Have Noises In Background , Change img_blur Value(1) and Threshold For Best Result
img_blur = cv.medianBlur(image, 1)
_,img_thresh = cv.threshold(img_blur, 50, 255, cv.THRESH_BINARY)
```

```python
plt.imshow(image,cmap = 'gray')
```

```python
# Change Image Output To String
img_txt = pytesseract.image_to_string(img_thresh)
print(img_txt)
```

```python
text_boxes = pytesseract.image_to_boxes(image)
print(text_boxes)
```

```python
# Split Words Where We have Space in Text
text_boxes_list = text_boxes.split('\n')
```

```python
# Read And Find Image Size
image_rgb = cv.imread("Image Address")
w,h,c = image_rgb.shape

# Find Coordinate Of Each Letters and Draw Rectangle around It
for box_coords in text_boxes_list:
    box_coords = box_coords.split(' ')
    if box_coords[0]:
       #print(box_coords)
        x1 = int(box_coords[1])
        y1 = int(box_coords[2])
        x2 = int(box_coords[3])
        y2 = int(box_coords[4])
        cv.rectangle(image_rgb, (x1,w-y1),(x2,w-y2),(255,0,0),2)
    plt.imshow(image_rgb)
```

```python
# Get Output Of Text as Dictionary
img_data = pytesseract.image_to_data(image, output_type=Output.DICT)
img_data
```

```python
# Find Each Words And Draw Rectangle around It
image_rgb = cv.imread("Image Address")
w,h,c = image_rgb.shape
for index,text in enumerate(img_data['text']):
    if text:
        x1 = int(img_data['left'][index])
        y1 = int(img_data['top'][index])
        x2 = int(img_data['width'][index])
        y2 = int(img_data['height'][index])
        cv.rectangle(image_rgb, (x1,y1),(x2,y2),(255,0,0),2)
plt.imshow(image_rgb)
```
